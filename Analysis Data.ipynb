```python
# Select the chosen features (replace with your final choices)
selected_features = ['scheme_name', 'returns_5yr', 'risk_level', 'returns_1yr', 'returns_3yr']

# Create a new DataFrame with the selected features
df_selected = df[selected_features]
```

```python
# Print the DataFrame with selected features for confirmation
df_selected
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Define features (replace with your final choices)
features = ['risk_level', 'returns_1yr', 'returns_3yr']
target = 'returns_5yr'
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_selected[features], df_selected[target], test_size=0.2, random_state=42)

# Create the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate model performance
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```

```python
print("R-squared:", r2)
print("Mean Squared Error:", mse)
```

```python
def categorize_risk(risk_level):
  """
  Categorizes risk level based on numerical value.

  Args:
      risk_level: Numerical risk level (1 to 6).

  Returns:
      Categorized risk level (low, medium, high).
  """
  if risk_level <= 2:
    return "low"
  elif risk_level <= 4:
    return "medium"
  else:
    return "high"
```

```python
df_selected["risk_category"] = df_selected["risk_level"].apply(categorize_risk)
```

```python
def recommend_funds(user_risk_tolerance, df_selected):
  """
  Recommends mutual funds based on user risk tolerance and returns (if 'predicted_return' is missing).

  Args:
      user_risk_tolerance: User's risk tolerance (low, medium, high).
      df_selected: DataFrame containing mutual fund data (without 'predicted_return').

  Returns:
      A list of recommended mutual funds.
  """

  if user_risk_tolerance == "high":
    sorted_df = df_selected.sort_values(by='returns_5yr', ascending=False)  # Prioritize potentially higher returns
  else:
    sorted_df = df_selected.sort_values(by='returns_5yr', ascending=True)   # Prioritize potentially lower returns

  # Select top N recommendations (optional)
  top_n_funds = sorted_df.head(n=5)  # Adjust number of recommendations
  return top_n_funds["scheme_name"].tolist()

# Example Usage
user_risk_tolerance = input("Enter your risk tolerance (low, medium, high): ")
recommended_funds = recommend_funds(user_risk_tolerance, df_selected.copy())
print("Recommended Funds for You:", recommended_funds)
```
